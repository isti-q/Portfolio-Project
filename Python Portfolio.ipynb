```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\ASUS\Downloads\movies.csv')
```

```python
#look some first part of the table
df.head()
```

```python
df
```

```python
#checking the type of the each columns of the data

df.dtypes
```

```python
#change the budget and gross data type from float(decimal) to int(integer)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
#some value on the year released columns and the year from released colom aren't same
#make the yearcorrect columns

df['released'] = df['released'].astype(str)
df['yearcorrect'] = df['released'].str[:4]

df
```

```python
#change the year column values with yearcorrect values and delete the yearcorrect column

df['year'] = df['yearcorrect']
del df['yearcorrect']
df
```

```python
#drop duplicate rows if there's any

df.drop_duplicates()
```

```python
#order the data by the gross

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#plot gross based on the budget

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Gross Earning vs Budget')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
#regression plot on gross and score

sns.regplot(x=df['score'], y=df['gross'])
plt.title('Gross vs Score')
plt.show()
```

```python
#plot the gross based on the rating

sns.stripplot(x=df['rating'], y=df['gross'])
plt.title('Gross based on the rating')
```

```python
#total gross that companies acieved

CompanyGrossSum = df.groupby(['company'])[["gross"]].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company'], ascending = False)
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 
CompanyGrossSumSorted
```

```python
#Correlation matrix between all numeric values
df.corr(method =  'pearson')
```

```python
#plot correlation matrix

matrix_correlation = df.corr()

sns.heatmap(matrix_correlation, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#numerized the object data type
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
#Correlation matrix
matrix_correlation = df_numerized.corr()

sns.heatmap(matrix_correlation, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#show the each attributes correlations

correlation_pairs = df.corr().unstack()

#pd.set_option('display.max_rows', None)

correlation_pairs
```

```python
#sort the correlation pairs

sorted_pairs = correlation_pairs.sort_values()

sorted_pairs
```

```python
#show data for the correlation more than 0.5

high_correlation = sorted_pairs [abs(sorted_pairs)>0.5]

high_correlation
```
